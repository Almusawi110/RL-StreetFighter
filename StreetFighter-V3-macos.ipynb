# Hyperparameter tune

```python
# Importing the optimzation frame - HPO
import optuna
# PPO algo for RL
from stable_baselines3 import PPO
# Bring in the eval policy method for metric calculation
from stable_baselines3.common.evaluation import evaluate_policy
# Import the sb3 monitor for logging 
from stable_baselines3.common.monitor import Monitor
# Import the vec wrappers to vectorize and frame stack
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack
# Import os to deal with filepaths
import os
```

```python
LOG_DIR = './logs/'
OPT_DIR = './opt-models/'
```

```python
# Function to return test hyperparameters - define the object function
def optimize_ppo(trial): 
   return {
        'n_steps': trial.suggest_int('n_steps', 2048, 8192),
        'gamma': trial.suggest_loguniform('gamma', 0.8, 0.9999),
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-5, 1e-4)
    }
```

```python
SAVE_PATH = os.path.join(OPT_DIR, 'trial_{}_best_model'.format(1))
```

```python
import tensorflow as tf

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("Memory growth enabled for GPU.")
    except RuntimeError as e:
        print(e)
```

```python
# Run a training loop and return mean reward 
from stable_baselines3.common.vec_env import SubprocVecEnv, VecFrameStack
import torch

# Helper function to create environments
def make_env():
    """
    Returns a callable to create an instance of the StreetFighter environment.
    SubprocVecEnv requires environments to be callable.
    """
    return lambda: Monitor(StreetFighter(), LOG_DIR)

def optimize_agent(trial):
    try:
        model_params = optimize_ppo(trial)

        # Multi-threaded environment setup
        num_envs = 4  # Number of parallel environments
        env = SubprocVecEnv([make_env() for _ in range(num_envs)])  # Multi-threaded environments
        env = VecFrameStack(env, 4, channels_order='last')  # Frame stacking

        # Detect device (MPS or CPU)
        device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

        # Create PPO model
        model = PPO(
            'CnnPolicy',
            env,
            tensorboard_log=LOG_DIR,
            verbose=0,
            device=device,  # Use MPS or CPU
            **model_params
        )
        model.learn(total_timesteps=100000)

        # Evaluate the model
        mean_reward, _ = evaluate_policy(model, env, n_eval_episodes=5)
        env.close()

        # Save the model
        SAVE_PATH = os.path.join(OPT_DIR, f'trial_{trial.number}_best_model')
        model.save(SAVE_PATH)

        return mean_reward

    except Exception as e:
        print(f"Error during trial {trial.number}: {e}")
        return -1000
```

```python
env.close()
```

```python
import optuna

# Storage for distributed optimization
sqlite_url = "sqlite:///optuna_study.db"

# Create study with advanced sampler and pruning
study = optuna.create_study(
    direction='maximize',
    storage=sqlite_url,
    load_if_exists=True,
    sampler=optuna.samplers.TPESampler(multivariate=True),
    pruner=optuna.pruners.MedianPruner()
)

# Optimize with parallel trials
study.optimize(optimize_agent, n_trials=60, n_jobs=8)
```

# Setup Callback

```python
# Import base callback 
from stable_baselines3.common.callbacks import BaseCallback
```

```python
class TrainAndLoggingCallback(BaseCallback):

    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'best_model_{}'.format(self.n_calls))
            self.model.save(model_path)

        return True
```

```python
CHECKPOINT_DIR = './train/'
```

```python
callback = TrainAndLoggingCallback(check_freq=10000, save_path=CHECKPOINT_DIR)
```
